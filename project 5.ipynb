# Scientific Experiment Data Analysis with NumPy

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  
temperature_data = np.random.normal(25, 5, 100)

mean_temperature = np.mean(temperature_data)
median_temperature = np.median(temperature_data)
std_deviation = np.std(temperature_data)
variance = np.var(temperature_data)

print(f"Mean Temperature: {mean_temperature:.2f} 째C")
print(f"Median Temperature: {median_temperature:.2f} 째C")
print(f"Standard Deviation: {std_deviation:.2f} 째C")
print(f"Variance: {variance:.2f} 째C")

plt.figure(figsize=(8, 5))  

plt.hist(temperature_data, bins=15, edgecolor='black', color='skyblue')
plt.xlabel('Temperature (째C)')
plt.ylabel('Frequency')
plt.title('Temperature Measurement Distribution')


plt.show()
```
