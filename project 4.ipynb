# Movie Ratings Data Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

#Load a sample movie ratings dataset into a pandas DataFrame
data = {
    'Movie': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E'],
    'Rating': [4.5, 4.2, 3.8, 4.0, 4.7],
    'Genre': ['Drama', 'Action', 'Comedy', 'Drama', 'Sci-Fi'],
    'Year': [2019, 2020, 2018, 2019, 2021]
}

df = pd.DataFrame(data)

#  Explore and analyze the dataset
average_rating = df['Rating'].mean()
top_rated_movies = df[df['Rating'] == df['Rating'].max()]

print(f"Average Rating: {average_rating}")
print("Top-Rated Movies:")
print(top_rated_movies)
#Create visualizations
plt.figure(figsize=(8, 5)) 

plt.bar(df['Movie'], df['Rating'], color='lightblue')
plt.xlabel('Movie')
plt.ylabel('Rating')
plt.title('Movie Ratings')

# Display the bar chart
plt.xticks(rotation=15)  
plt.show()
```
