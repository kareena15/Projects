```python
import matplotlib.pyplot as plt

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
fruit_counts = [35, 42, 15, 8]
colors=['blue','green','grey','red']

plt.figure(figsize=(6, 6))  

explode = (0.1, 0, 0, 0)

plt.title('Fruit Distribution')

plt.pie(fruit_counts, labels=fruits, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)

plt.legend(fruits, loc='best')

plt.show
```
